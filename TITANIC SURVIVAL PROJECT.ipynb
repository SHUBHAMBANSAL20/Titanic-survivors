```python
X_train = train.iloc[:,[2,4,5]].values
y_train = train.iloc[:,1].values
```

```python
X_train
```

```python
X_test = test.iloc[:,[1,3,4]].values

"""age missing values changing """
from sklearn.preprocessing import Imputer 
imputer = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
imputer = imputer.fit(X_train[:,2:3])
X_train[:,2:3] = imputer.transform(X_train[:,2:3])

""" sex changing"""
from sklearn.preprocessing import LabelEncoder
labelencoder_x = LabelEncoder()
X_train[:,1] = labelencoder_x.fit_transform(X_train[:,1])

""" p_class changing"""
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoders_x = LabelEncoder()
X_train[:,0] = labelencoders_x.fit_transform(X_train[:,0])

onehotencoder = OneHotEncoder(categorical_features = [0])
X_train = onehotencoder.fit_transform(X_train).toarray()

"""age missing in test"""
from sklearn.preprocessing import Imputer 
imputers = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
imputers = imputers.fit(X_test[:,2:3])
X_test[:,2:3] = imputers.transform(X_test[:,2:3])


"""sex changing of test dataset"""
from sklearn.preprocessing import LabelEncoder
labelencode_x = LabelEncoder()
X_test[:,1] = labelencode_x.fit_transform(X_test[:,1])


"""p_class changing of test dataset"""
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoderss_x = LabelEncoder()
X_test[:,0] = labelencoderss_x.fit_transform(X_test[:,0])

onehotencode = OneHotEncoder(categorical_features = [0])
X_test = onehotencode.fit_transform(X_test).toarray()

"""fitting of model"""
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

#prediction

y_pred = classifier.predict(X_test)


```
